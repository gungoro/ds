```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.linear_model import LinearRegression
import seaborn as sns
%matplotlib inline
```

```python
d = datasets.california_housing.fetch_california_housing()
```

```python
X = d.data
y = d.target
```

```python
X_corr = np.corrcoef(X.transpose())
# masking the upper triangular part
mask = np.zeros_like(X_corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(X_corr, mask=mask, xticklabels=d.feature_names, yticklabels=d.feature_names)
```
